
# Arithmetic Operators



These are symbols that represent an arithmetic computation. For example, the + performs addition.

```python
55 + 12
```

```python
# The - sign is for subtraction
79 - 12
```

```python
# The * symbol is to perform Multiplication
6 * 7
```

```python
# The / symbol is used for division
134 / 2
```

### When you divide in python, the answer will always be a data type called a float. 

### Integer - represents numbers with no fractional or decimal part

### Floating-point numbers - represent numbers with a decimal or fractional value

```python
# the // symbol stands for Floor Division (Integer Division) 
84 // 2
```

### FLoor Division gives you everything before the decimal

```python
85 / 2
# everything from the decimal on gets removed fro our answer.
```

```python
85 // 2
```

```python
# The ** symbol is used to represent exponents
7 ** 2
```

```python
10 ** -325
```

```python
# The ^ symbol (the caret), in Python it is a bitwise operator called XOR
7^2
```

```python
# The % symbol is used for whaat we call Modulus Division
85 % 2
```
